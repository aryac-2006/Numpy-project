```python
import numpy as np

# 1D array of 20 random integers between 1 and 50
arr = np.random.randint(1, 51, 20)
print("1D Array: ")
print(arr)

# Reshape the array into a 4x5 matrix
matrix = arr.reshape(4, 5)
print("\nReshaped Matrix: ")
print(matrix)

# Calculate the sum, mean, standard deviation, and maximum value of the matrix
sum = matrix.sum()
print("\nSum: ", sum)

mean = matrix.mean()
print("Mean: ", mean)

std_dev = matrix.std()
print("Standard Deviation: ", std_dev)

row_max = matrix.max()
print("Maximum Value in each row: ", row_max)
```
