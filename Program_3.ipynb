```python
import numpy as np

# Numbers from 0 to 20 (step 1)
arr1 = np.arange(0, 21, 1)
print("a). Numbers from 0 to 20 (step 1):")
print(arr1)

# Even numbers from 10 to 50
arr2 = np.arange(10, 51, 2)
print("\nb). Even numbers from 10 to 50:")
print(arr2)

# Numbers from 5 to 100 with step of 5
arr3 = np.arange(5, 101, 5)
print("\nc). Numbers from 5 to 100 with step of 5:")
print(arr3)
```
