```python
import numpy as np

v1 = np.array([ 2, 4, 6, 8 ])
v2 = np.array([ 1, 3, 5, 7 ])

# Create vectors
print("Array 1:", v1)
print("Array 2:", v2)

# Addition
print("\nAddition:")
a = v1 + v2
print(a)

# Subtraction
print("\nSubtraction:")
s = v1 - v2
print(s)

# Multiplication
print("\nElement-wise Multiplication:")
m = v1 * v2
print(m)

# Dot Product
print("\nDot Product:")
d = np.dot(v1, v2)
print(d)
```
