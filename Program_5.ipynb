```python
import numpy as np

#  A 1D array of 10 random numbers between 0 and 1 using np.random.rand()
arr1 = np.random.rand(10)
print("\na). 1D array of 10 random numbers between 0 and 1:")
print(" ", arr1)

#  A 3x3 matrix of random numbers from standard normal distribution using np.random.randn()
arr2 = np.random.randn(3, 3)
print("\nb). 3x3 matrix of random numbers from standard normal distribution:")
print(" ", arr2)

# A 2D array 4x5 of random integers between 10 and 50 using np.random.randint()
arr3 = np.random.randint(10, 51, (4, 5))
print("\nc). 2D array of shape 4x5 of random integers between 10 and 50:")
print(" ", arr3)
```
